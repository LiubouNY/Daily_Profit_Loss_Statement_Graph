```python
import pandas as pd 
from matplotlib import pyplot as plt
import plotly.express as px
```

```python
#open statement file in csv format from local disc
df = pd.read_csv(r'C:\Users\16469\P&L_statements\2024-03-15-AccountStatement_streamlit.csv')
df.head(15)
```

```python
#making third row values as column headers
df.columns = df.iloc[2]
df.head()
```

```python
#shifting row up, removing empty rows
df = df.shift(-3)
df.head(10)
```

```python
#selecting only needed columns and renaming columns
needed_columns = ['TIME', 'AMOUNT', 'BALANCE']
df = df[needed_columns]
df.index.names = ['INDEX']
df.head(10)
```

```python
#craeting simple line chart plotly.express 
px.line(df, x = 'TIME', y = 'BALANCE', title = 'Daily Balance Fluctuation')
```
