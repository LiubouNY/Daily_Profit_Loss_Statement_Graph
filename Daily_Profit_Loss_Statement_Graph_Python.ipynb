```python
import pandas as pd 
from matplotlib import pyplot as plt
import plotly.express as px
import mplcyberpunk
```

```python
#open statement file in csv format from local disc
df = pd.read_csv(r'C:\Users\16469\P&L_statements\2024-03-15-AccountStatement_streamlit.csv')
df.head(15)
```

# Data Cleaning

```python
#making third row values as column headers
df.columns = df.iloc[2]
df.head()
```

```python
#shifting row up, removing empty rows
df = df.shift(-3)
df.head(10)
```

```python
#selecting only needed columns and renaming columns
needed_columns = ['TIME', 'AMOUNT', 'BALANCE']
df = df[needed_columns]
df.index.names = ['INDEX']
df.head(10)
```

Converting Data Frame data types to craete chart

```python
#taking a look at the data dimentions
df.shape
```

```python
#taking a look at currents data types
df.dtypes
```

```python
#converting current object data types to date and numeric types
df['TIME'] = pd.to_datetime(df['TIME'])
df['BALANCE'] = df['BALANCE'].str.replace(',','')
df['BALANCE'] = df['BALANCE'].str.replace('$','')
df['BALANCE'] = df['BALANCE'].astype(float)
df['AMOUNT'] = df['AMOUNT'].str.replace(',','')
df['AMOUNT'] = df['AMOUNT'].str.replace('$','')
df['AMOUNT'] = df['AMOUNT'].astype(float)
df.head()
```

```python
#check data types
df.dtypes
```

```python
#using mplcyberpunk library to create more detailed chart
# create data - using two measure, create one line chart
x = df['TIME'] 
y = df['BALANCE'] 

# Create the plot
plt.style.use('cyberpunk')
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(x, y, color="#4dd2ff", linewidth=2)
ax.set_title("P&L Statement. Daily Balace Fluctuation")
ax.set_xlabel("TIME")
ax.set_ylabel("BALANCE")


# Add Cyberpunk visual elements
mplcyberpunk.add_glow_effects()

# Show the plot
plt.show()
```

```python
#seting up data for chart
x = df['TIME'] 
y = df['AMOUNT'] 


# Create the plot
plt.style.use('cyberpunk')
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(x, y, color='lime', linewidth=2)

ax.set_title("P&L Statement. Daily Amount Fluctuation")
ax.set_xlabel("TIME")
ax.set_ylabel("AMOUNT")


# add glowing effect
mplcyberpunk.add_glow_effects()

# Show the plot
plt.show()
```
